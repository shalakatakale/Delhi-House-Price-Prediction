# Delhi House Price Prediction using Machine Learning

```python
import numpy as np
import pandas as pd
```

```python
housing = pd.read_csv("ReadytoUse.csv")
```

```python
housing.head()
```

```python
# Split X and y
y = housing["Price"]
X = housing.drop("Price",axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1,random_state=42)
print(f"Rows in train set: {len(X_train)} \nRows in test set: {len(X_test)}\n")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = pd.DataFrame(scaler.transform(X_train), index=X_train.index, columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), index=X_test.index, columns=X_test.columns)
```

```python
X_train.head()
```

## Traing Machine Learning Models

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
lr_model = LinearRegression()
ridge_model = Ridge()
lasso_model = Lasso()
knn_model = KNeighborsRegressor()
dtr_model = DecisionTreeRegressor()
rf_model = RandomForestRegressor()
xgb_model = XGBRegressor()
```

```python
lr_model.fit(X_train,y_train)
ridge_model.fit(X_train,y_train)
lasso_model.fit(X_train,y_train)
knn_model.fit(X_train,y_train)
dtr_model.fit(X_train,y_train)
rf_model.fit(X_train,y_train)
xgb_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
lr_predictions = lr_model.predict(X_test)
lr_mse = mean_squared_error(y_test,lr_predictions)
lr_rmse = np.sqrt(lr_mse)

ridge_predictions = ridge_model.predict(X_test)
ridge_mse = mean_squared_error(y_test,ridge_predictions)
ridge_rmse = np.sqrt(ridge_mse)

lasso_predictions = lasso_model.predict(X_test)
lasso_mse = mean_squared_error(y_test,lasso_predictions)
lasso_rmse = np.sqrt(lasso_mse)

knn_predictions = knn_model.predict(X_test)
knn_mse = mean_squared_error(y_test,knn_predictions)
knn_rmse = np.sqrt(knn_mse)

dtr_predictions = dtr_model.predict(X_test)
dtr_mse = mean_squared_error(y_test,dtr_predictions)
dtr_rmse = np.sqrt(dtr_mse)

rf_predictions = rf_model.predict(X_test)
rf_mse = mean_squared_error(y_test,rf_predictions)
rf_rmse = np.sqrt(rf_mse)

xgb_predictions = xgb_model.predict(X_test)
xgb_mse = mean_squared_error(y_test,xgb_predictions)
xgb_rmse = np.sqrt(xgb_mse)

Scores_df = pd.DataFrame([{'Model': 'Linear Regression','Score':lr_model.score(X_test, y_test), "RMSE":lr_rmse},
                    {'Model': 'Ridge Regression','Score':ridge_model.score(X_test, y_test), "RMSE":ridge_rmse},
                    {'Model': 'Lasso','Score':lasso_model.score(X_test, y_test), "RMSE":lasso_rmse},
                    {'Model': 'KNN','Score':knn_model.score(X_test, y_test), "RMSE":knn_rmse},
                    {'Model': 'DTR','Score':dtr_model.score(X_test, y_test), "RMSE":dtr_rmse},
                    {'Model': 'Random Forest','Score':rf_model.score(X_test, y_test), "RMSE":rf_rmse},
                    {'Model': 'XGBoost','Score':xgb_model.score(X_test, y_test), "RMSE":xgb_rmse}],
                     columns=['Model','Score','RMSE'])

print(Scores_df)
```

### Cross Validation

```python
from sklearn.model_selection import cross_val_score
lr_scores = cross_val_score(lr_model,X_train,y_train,scoring="neg_mean_squared_error",cv =10)
lr_rmse_scores = np.sqrt(-lr_scores)

ridge_scores = cross_val_score(ridge_model,X_train,y_train,scoring="neg_mean_squared_error",cv =10)
ridge_rmse_scores = np.sqrt(-ridge_scores)

lasso_scores = cross_val_score(lasso_model,X_train,y_train,scoring="neg_mean_squared_error",cv =10)
lasso_rmse_scores = np.sqrt(-lasso_scores)

knn_scores = cross_val_score(knn_model,X_train,y_train,scoring="neg_mean_squared_error",cv =10)
knn_rmse_scores = np.sqrt(-knn_scores)

dtr_scores = cross_val_score(dtr_model,X_train,y_train,scoring="neg_mean_squared_error",cv =10)
dtr_rmse_scores = np.sqrt(-dtr_scores)

rf_scores = cross_val_score(rf_model,X_train,y_train,scoring="neg_mean_squared_error",cv =10)
rf_rmse_scores = np.sqrt(-rf_scores)

xgb_scores = cross_val_score(xgb_model,X_train,y_train,scoring="neg_mean_squared_error",cv =10)
xgb_rmse_scores = np.sqrt(-xgb_scores)
```

```python
def print_scores(scores):
    print("RMSE Scores:",scores)
    print("Mean:",scores.mean())
    print("Standard Deviation:",scores.std())    
    print("\n")
```

```python
# using lr_scores
print("Linear Regression")
print_scores(lr_rmse_scores)

# using ridge_scores
print("Ridge Regression")
print_scores(ridge_rmse_scores)

# using lasso_scores
print("Lasso Regression")
print_scores(lasso_rmse_scores)

# using knn_scores
print("KNN")
print_scores(knn_rmse_scores)

# using dtr_scores
print("Decision Tree Regressor")
print_scores(dtr_rmse_scores)

# using rf_scores
print("Randome Forest Regressor")
print_scores(rf_rmse_scores)

# using xgb_scores
print("XGB Regression")
print_scores(xgb_rmse_scores)
```

## Test Model

```python
list(X.columns)
```

```python
def predict_house_price(model,BHK,Bathroom,Parking,Area,Per_Sqft,Furnishing,Status,Transaction,Type,Locality):
    x =np.zeros(len(X.columns)) 
    #print(x)
    x[0]=BHK
    x[1]=Bathroom
    x[2]=Parking
    x[3]=Area
    x[4]=Per_Sqft
    
    if "Furnishing" == "Semi-Furnished":
        x[5] = 1
    elif "Furnishing" == "Unfurnished":
        x[5] = 2
    elif "Furnishing" == "Furnished":
        x[5] = 0
        
    if "Status" == "Ready_to_move":
        x[6] = 1
    elif "Status" == "Almost_ready":
        x[6] = 0
    
    if "Transaction" == "New_Property":
        x[7] = 0
    elif "Transction" == "Resale":
        x[7] = 1
    
    if "Type" == "Builder_Floor":
        x[8] = 1
    elif "Type" == "Apartment":
        x[8] = 0
        
    if "Locality_"+Locality in X.columns:
        Locality_index = np.where(X.columns=="Locality_"+Locality)[0][0]
        x[Locality_index] =1
        
        
    # feature scaling
    x = scaler.transform([x]) # give 2d np array for feature scaling and get 1d scaled np array
    return model.predict(x)[0]
```

```python
predict_house_price(model=xgb_model,BHK=2,Bathroom=2,Parking=2,Area=1000,Per_Sqft=8667,Furnishing="Unfurnished",Status="Ready_to_move",Transaction="Resale",Type="Apartment",Locality="New Friends Colony")
```
